```python
features = ["Age", "Gender", "CreditScore", "Geography", "EstimatedSalary", "Tenure", "Balance", "NumOfProducts", "HasCrCard", "IsActiveMember"]

# Iterando sobre cada caracter√≠stica e gerando um gr√°fico
for feature in features:
    grouped_data = df.groupby(feature).agg({"Exited": ["count"]})
    
    # Plotando o gr√°fico de dispers√£o para a caracter√≠stica atual
    plt.scatter(grouped_data.index, grouped_data[('Exited', 'count')], s=20)
    
    # Adicionando t√≠tulo e r√≥tulos aos eixos
    plt.title(f'Churn X {feature}')
    plt.xlabel(feature)
    plt.ylabel('Churn - contagem total')
    
    # Exibindo o gr√°fico
    plt.show()
```

```python
features_numericas = ["Age", "CreditScore",  "EstimatedSalary", "Tenure", "Balance"]

# Definir o n√∫mero m√°ximo de colunas por linha
max_cols = 3

# Determinar o n√∫mero de linhas necess√°rias sem usar a biblioteca math
n_rows = len(features_numericas) // max_cols + (1 if len(features_numericas) % max_cols != 0 else 0)

# Criando os subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=max_cols, figsize=(15, 5 * n_rows))

# Achatar a grade de eixos para f√°cil indexa√ß√£o (caso haja mais de uma linha)
axes = axes.flatten()

# Iterando sobre cada caracter√≠stica e gerando um gr√°fico em um subplot
for i, feature in enumerate(features_numericas):
    grouped_data = df.groupby(feature).agg({"Exited": ["count"]})
    
    # Plotando o gr√°fico de dispers√£o no subplot espec√≠fico
    axes[i].scatter(grouped_data.index, grouped_data[('Exited', 'count')], s=20)
    
    # Adicionando t√≠tulo e r√≥tulos aos eixos
    axes[i].set_title(f'Churn X {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Churn - contagem total')

# Remover subplots vazios, se houver
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Ajustando o layout para evitar sobreposi√ß√£o
plt.tight_layout()

# Exibindo os gr√°ficos
plt.show()
```
